## What and why <a name="what_and_why"></a>

**What**: Integration is a mathematical technique under the field of calculus that for our purposes will be used to find the area under curves.

**Why**: We need integration because we want to find quantities like $\mathbb{P}(X \leq 2)$ given $X$ is a continuous RV.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def normal_density(x):
    return np.exp(-x**2/2)/np.sqrt(2*np.pi)
    
x = np.linspace(-5,5,100)
plt.plot(x,normal_density(x))
plt.xlabel('x')
plt.ylabel('density f(x)')
plt.title('Normal density')
plt.vlines(2,0,normal_density(2))
plt.show()
```

## Theory of integration <a name="theory"></a>

The idea is to partition the domain into small intervals. In these interval the function is approximated by a constant. Thus, we are left with series of retangles which is much easier to compute.

```python
plt.plot(x,normal_density(x))
plt.xlabel('x')
plt.ylabel('density f(x)')
plt.title('Normal density')
plt.vlines(2,0,normal_density(2))
plt.show()
```

```python
partition = np.linspace(-5,2,20)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
for i in range(len(partition)-1):
    ax.plot([partition[i],partition[i]],[0,normal_density(partition[i])],'k')
    ax.hlines(normal_density(partition[i]),partition[i],partition[i+1],'k')
    if partition[i]>0:
        ax.plot([partition[i+1],partition[i+1]],[0,normal_density(partition[i])],'k')
        
plt.plot(x,normal_density(x),linewidth=4)
plt.show()
```

## Common integrals <a name="common"></a>

This is not a calculus course so I am not going to prove any results. Instead I will just state some common integrals.
* $\int x^n dx = \frac{1}{n+1}x^{n+1} + c$
* $\int e^{bx} dx = \frac{1}{b}e^{bx}+ c$

I think two is enough for now.

Above are what you call indefinite integrals. In order to find the area under a curve we need to use definite integrals.

* $\int_{c}^{d} x^n dx = \frac{1}{n+1}d^{n+1} - \frac{1}{n+1}c^{n+1}$
* $\int_{c}^{d} e^{bx} dx = \frac{1}{b}e^{bd} - \frac{1}{b}e^{bc}$

Integrals are linear this means you can add them together and multiply them by scalars.


* $\int f(x) dx + \int g(x) dx = \int( f(x)+  g(x) )dx$
* $a \int f(x) dx  = \int af(x) dx$

Great lets have an example.

Evaluate $\int_{2}^{1} e^{10x}$

$$\int_{1}^{2} e^{10x} dx =\left[\frac{1}{10} e^{10x} \right]_{1}^{2}= \frac{1}{10}e^{20} - \frac{1}{10}e^{10}$$


Let's have one more example to show linearity.

Evaluate $\int_{2}^{1}( x^{2}+ 3x) dx$

$$\int_{1}^{2}( x^{2}+ 2x) dx =\int_{1}^{2} x^{2} dx + \int_{1}^{2} 2xdx =  \left[\frac{x^3}{3}\right]_{1}^{2} + \left[x^2\right]_{1}^{2} = \frac{8}{3} - \frac{1}{3} + 4 - 1 = \frac{16}{3}$$
